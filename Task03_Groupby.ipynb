```python
import pandas as pd
import numpy as np

# Đọc dữ liệu sạch
graduate = pd.read_csv('data_clean/graduate_info_clean.csv')
employment = pd.read_csv('data_clean/employment_survey_clean.csv')
certifications = pd.read_csv('data_clean/certifications_clean.csv')

# Merge employment với graduate
emp_grad = employment.merge(graduate[['student_id', 'major', 'grad_year']], 
                            on='student_id', how='left')

print("Dữ liệu được tải thành công!")
```

## 1. Mức lương khởi điểm trung bình theo ngành

```python
print("\n" + "="*80)
print("1. MỨC LƯƠNG TRUNG BÌNH THEO NGÀNH")
print("="*80)

# Lọc lương hợp lệ (> 0)
valid_salary = emp_grad[emp_grad['salary'] > 0]

salary_by_major = valid_salary.groupby('major').agg({
    'salary': ['count', 'mean', 'min', 'max', 'std']
}).round(0)

salary_by_major.columns = ['Số bản ghi', 'Lương TB', 'Min', 'Max', 'Std Dev']
salary_by_major = salary_by_major.astype(int)

print(salary_by_major)
print(f"\nNgành có lương TB cao nhất: {salary_by_major['Lương TB'].idxmax()}")
print(f"Ngành có lương TB thấp nhất: {salary_by_major['Lương TB'].idxmin()}")
```

## 2. Mức lương trung bình theo năm tốt nghiệp

```python
print("\n" + "="*80)
print("2. MỨC LƯƠNG TRUNG BÌNH THEO NĂM TỐT NGHIỆP")
print("="*80)

salary_by_year = valid_salary.groupby('grad_year').agg({
    'salary': ['count', 'mean', 'min', 'max']
}).round(0)

salary_by_year.columns = ['Số bản ghi', 'Lương TB', 'Min', 'Max']
salary_by_year = salary_by_year.astype(int)

print(salary_by_year)
print(f"\nNăm có lương TB cao nhất: {salary_by_year['Lương TB'].idxmax()}")
```
