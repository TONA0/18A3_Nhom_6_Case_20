# ĐỌC DỮ LIỆU FILE CLEAN

```python
import pandas as pd
import numpy as np

# Đọc dữ liệu sạch
graduate = pd.read_csv('data_clean/graduate_info_clean.csv')
employment = pd.read_csv('data_clean/employment_survey_clean.csv')
certifications = pd.read_csv('data_clean/certifications_clean.csv')

print(f"Graduate: {len(graduate)} hàng")
print(f"Employment: {len(employment)} hàng")
print(f"Certifications: {len(certifications)} hàng")
```

1. MERG GRADUATE VÀ EMPLOYMENT THEO student_id

```python
print("1. MERGE GRADUATE VÀ EMPLOYMENT")
print("="*80)

# dùng left để giữ sinh viên theo student_id
merged_emp = graduate.merge(employment, on='student_id', how='left')

print(f"\nKết quả merge: {len(merged_emp)} hàng") # đếm số hàng sau khi merg
print(f"Sinh viên có dữ liệu việc làm: {merged_emp['salary'].notna().sum()}") # dùng notna để xác dữ liệu sv có việc làm, true nếu có dữ liệu, false nếu là NaN
print(f"Sinh viên KHÔNG có dữ liệu việc làm: {merged_emp['salary'].isna().sum()}") # dùng isna xdinh sv kh có dữ liệu việc làm, true nếu giá trị NaN

print("\nSample kết quả merge:")
print(merged_emp[['student_id', 'full_name', 'major', 'grad_year', 'job_title', 'salary']].head(5))
```

2. MERG THÊM BẢNG certifications

```python
print("2. MERGE THÊM CERTIFICATIONS")
print("="*80)

# Đếm số chứng chỉ certifications
cert_count = certifications.groupby('student_id').size().reset_index(name='cert_count')

# Merge certification với merged_emp
final_merged = merged_emp.merge(cert_count, on='student_id', how='left')
final_merged['cert_count'] = final_merged['cert_count'].fillna(0).astype(int)

print(f"\nKết quả merge cuối cùng: {len(final_merged)} hàng")
print(f"Cột trong merged: {final_merged.columns.tolist()}")

print("\nSample final merged:")
print(final_merged[['student_id', 'full_name', 'major', 'job_title', 'salary', 'cert_count']].head(5))
```
